```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("T:\Machine Learning Course\Machine Learning\DataSet\placement.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
input_data = df[['cgpa']]
output_data = df['package']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.20, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict([[6.89]])
```

```python
lr.score(x_test, y_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(input_data)
```

```python
sns.scatterplot(x="cgpa",y="package",data=df)
plt.plot(df['cgpa'], y_pred, c='red')
plt.legend(["Orginal data", "Predict Line"])
plt.savefig(r"T:/Machine Learning Course/Machine Learning/Figures/Predict.jpg")
plt.show()
```
