```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df =  pd.read_csv("Dataset/loan_data_set.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
mm.fit(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_mm"] = pd.DataFrame(mm.transform(df[["CoapplicantIncome"]]))
```

```python
df
```

```python
df.describe()
```

```python
plt.subplot(1, 2, 1)
sns.distplot(df["CoapplicantIncome"])
plt.title("Before")
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df["CoapplicantIncome_mm"])
plt.show()
```
